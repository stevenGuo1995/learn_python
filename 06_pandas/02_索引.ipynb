# Pandas 索引结构

```python
import pandas as pd
df = pd.read_csv('./data/titanic.csv')
```

```python
# 定位到某一列
df['Age'][:5]
```

```python
# 定位到多个列, 注意需要两个中括号
df[['Age', 'Fare']][:5]
```

## 定位到具体数据
* loc 用label定位
* iloc 用position定位

```python
df.iloc[0]
```

```python
df.iloc[:5]
```

```python
df.iloc[0:5, 1:3]
```

```python
# 设置索引列
df = df.set_index('Name')
```

```python
df.loc['Braund, Mr. Owen Harris']
```

```python
df.loc['Braund, Mr. Owen Harris','Age']
```

```python
# 字符串索引也可以进行切片
df.loc['Braund, Mr. Owen Harris':'Allen, Mr. William Henry']
```

```python
# 对某个值进行复制
df.loc['Braund, Mr. Owen Harris','Age'] = 20
```

```python
df.loc['Braund, Mr. Owen Harris','Age']
```

## Bool类型索引

```python
df['Fare'] > 40
```

```python
# 获取所有Fare大于40的行
df[df['Fare'] > 40][:5]
```

```python
# 计算所有男性的平均年龄
df.loc[df['Sex'] == 'male','Age'].mean()
```
