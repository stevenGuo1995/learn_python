```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def sin_plot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5 ) * (7 - i) * flip)
```

```python
sin_plot()
```

```python
# seaborn 默认风格
sns.set()
sin_plot()
```

## 5种主题风格
* darkgrid
* whitegrid
* dark
* white
* ticks

```python
sns.set_style('whitegrid')
sin_plot()
```

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style('dark')
sin_plot()
```

```python
sns.set_style('white')
sin_plot()
```

```python
sns.set_style('ticks')
sin_plot()
```

```python
sin_plot()

# 去掉某些轴
sns.despine()
```

```python
sin_plot()

# 参数offset————离轴的距离
sns.despine(offset=100)
```

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
sns.despine(left=False)
```

```python
# with内是一种风格，外面是另一种风格
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
    sns.boxplot(data=data)

plt.subplot(212)
sns.boxplot(data=data)
```

```python
sns.set()
```
