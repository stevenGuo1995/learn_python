```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10)
y = np.sin(x)

plt.plot(x, y)
```

## 子图操作

```python
# 211表示要画的图是两行一列的图，最后一个1表示第一个图
plt.subplot(211)
plt.plot(x, y, 'r')

plt.subplot(212)
plt.plot(x, y, 'g')
```

```python
plt.subplot(121)
plt.plot(x, y, 'r')

plt.subplot(122)
plt.plot(x, y, 'g')
```

```python
plt.subplot(221)
plt.plot(x, y, 'r')

plt.subplot(224)
plt.plot(x, y, 'g')
```

## 添加标注

```python
plt.plot(x, y, color='b', linestyle=':',marker='o',markerfacecolor='r', markersize=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin')
plt.text(0,0, 'zero')
plt.grid(True)
plt.annotate('note',xy=(-5,1),xytext=(-2.5,0.5), arrowprops=dict(facecolor='black', shrink=0.05))
```
