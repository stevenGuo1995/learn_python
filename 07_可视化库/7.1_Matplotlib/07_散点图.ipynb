```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu_vecl = np.array([0, 0]) # 均值
cov_matl = np.array([[2, 0], [0, 2]]) # 标准差

x1_samples = np.random.multivariate_normal(mu_vecl, cov_matl, 100)
x2_samples = np.random.multivariate_normal(mu_vecl + 0.2, cov_matl + 0.2, 100)
x3_samples = np.random.multivariate_normal(mu_vecl + 0.4, cov_matl + 0.4, 100)

plt.figure(figsize=(8, 6))
plt.scatter(x1_samples[:,0], x1_samples[:,1], marker='x', color='blue', label='x1')
plt.scatter(x2_samples[:,0], x2_samples[:,1], marker='o', color='red', label='x2')
plt.scatter(x3_samples[:,0], x3_samples[:,1], marker='s', color='green', label='x3')

plt.legend()
```

```python
import random
# 添加坐标
x_coords = [random.random() for i in range(10)]
y_coords = [random.random() for i in range(10)]

plt.scatter(x_coords, y_coords)

for x, y in zip(x_coords, y_coords):
    plt.annotate('({:.2f}, {:.2f})'.format(x, y), xy=(x, y), xytext=(0,-15), textcoords='offset points', ha='center')
```
