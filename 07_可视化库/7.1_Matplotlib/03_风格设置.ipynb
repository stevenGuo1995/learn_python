```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# matplotlib的风格
plt.style.available
```

```python
x = np.linspace(-10, 10)
y = np.sin(x)
plt.plot(x, y)
```

```python
plt.style.use('bmh')
plt.plot(x, y)
```

```python
plt.style.use('ggplot')
plt.plot(x, y)
```

```python
plt.style.use('seaborn-deep')
plt.plot(x, y)
```

```python
# 风格混用
plt.style.use(['dark_background', 'seaborn-deep'])
plt.plot(x, y)
```

```python
plt.xkcd()
plt.plot(x, y)
```
